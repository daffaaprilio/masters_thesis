## Variant analysis: Creating sets of private genes (for confirmation of `bcftools isec`)

```python
def sites(df):
    return set(zip(df["chrom"], df["pos"], df["ref"], df["alt"]))

sites_sbc10 = sites(df_sbc10)
sites_sbc4  = sites(df_sbc4)
sites_sbc11 = sites(df_sbc11)
sites_sbc23 = sites(df_sbc23)
```

```python
sample_sites = {
    "SBC4":  sites_sbc4,
    "SBC10": sites_sbc10,
    "SBC11": sites_sbc11,
    "SBC23": sites_sbc23,
}

for name, s in sample_sites.items():
    others = {k: v for k, v in sample_sites.items() if k != name}
    exclusive = s - set().union(*others.values())
    print(f"── {name} ──────────────────────────")
    for other_name, other_set in others.items():
        print(f"  {name} ∩ {other_name}: {len(s & other_set)}")
    print(f"  Exclusive to {name}: {len(exclusive)}")
    print(f"  Total in {name}:     {len(s)}")
    print()
```

```python
sites_sbc10_high     = sites(df_sbc10[df_sbc10["impact"] == "HIGH"])
sites_sbc10_moderate = sites(df_sbc10[df_sbc10["impact"] == "MODERATE"])
sites_sbc10_low      = sites(df_sbc10[df_sbc10["impact"] == "LOW"])
sites_sbc10_modifier = sites(df_sbc10[df_sbc10["impact"] == "MODIFIER"])

print(f"HIGH:     {len(sites_sbc10_high):,}")
print(f"MODERATE: {len(sites_sbc10_moderate):,}")
print(f"LOW:      {len(sites_sbc10_low):,}")
print(f"MODIFIER: {len(sites_sbc10_modifier):,}")
print(f"Total:    {len(sites_sbc10_high | sites_sbc10_moderate | sites_sbc10_low | sites_sbc10_modifier):,}")
```

```python
import re

_GENE_ID_RE = re.compile(r"SORBI_\d{4}G\d{6}")

def extract_gene_ids(series: pd.Series) -> list[str]:
    """Return sorted unique SORBI gene IDs extracted from a gene_name/gene_id column.

    Handles non-standard entries such as:
      CHR_START-SORBI_3001G000100
      SORBI_3001G000100-SORBI_3001G000200
    by scanning each cell for all regex matches of the canonical pattern.
    """
    ids: set[str] = set()
    for val in series.dropna():
        ids.update(_GENE_ID_RE.findall(str(val)))
    return sorted(ids)


# Example: unique genes with HIGH-impact variants in SBC10
high_genes = extract_gene_ids(df_sbc10[df_sbc10["impact"] == "HIGH"]["gene_name"])
print(f"Unique HIGH-impact genes in SBC10: {len(high_genes)}")
high_genes[:10]
```

```python
gene_info = pd.read_csv(
    WDIR / "resources/NCBI_FTP/gene_info_4558",
    sep="\t",
    usecols=["GeneID", "Symbol", "LocusTag"],
    dtype={"GeneID": int, "Symbol": str, "LocusTag": str},
)

# Keep only rows with a SORBI_ locus tag and build lookup dicts
_sorbi_rows = gene_info[gene_info["LocusTag"].str.startswith("SORBI_", na=False)]

_sorbi_to_egi: dict[str, int] = (
    _sorbi_rows.set_index("LocusTag")["GeneID"].to_dict()
)
_sorbi_to_symbol: dict[str, str] = (
    _sorbi_rows.set_index("LocusTag")["Symbol"].to_dict()
)

def sorbi_to_egi(gene_ids: list[str]) -> dict[str, int | None]:
    """Convert a list of SORBI_ gene IDs to NCBI Entrez Gene IDs (EGI).

    Returns a dict mapping each input ID to its EGI, or None if not found.
    """
    return {gid: _sorbi_to_egi.get(gid) for gid in gene_ids}


def sorbi_to_symbol(gene_ids: list[str]) -> dict[str, str | None]:
    """Convert a list of SORBI_ gene IDs to gene symbols (e.g. LOC8060885).

    Returns a dict mapping each input ID to its symbol, or None if not found.
    """
    return {gid: _sorbi_to_symbol.get(gid) for gid in gene_ids}


print(f"Conversion table loaded: {len(_sorbi_to_egi):,} SORBI → EGI mappings")
print(f"                         {len(_sorbi_to_symbol):,} SORBI → symbol mappings")

# Quick check on the HIGH-impact genes
egi_map    = sorbi_to_egi(high_genes)
symbol_map = sorbi_to_symbol(high_genes)
n_mapped   = sum(v is not None for v in egi_map.values())
print(f"HIGH-impact genes mapped: {n_mapped}/{len(high_genes)}")
list(zip(list(egi_map.items())[:5], list(symbol_map.values())[:5]))
```

```python
def write_egi(gene_ids: list[str], path: str | Path) -> None:
    """Write EGIs for the given SORBI gene IDs to a one-per-line text file.

    Skips genes with no EGI mapping and prints a summary.
    """
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)

    egi_map  = sorbi_to_egi(gene_ids)
    mapped   = [egi for egi in egi_map.values() if egi is not None]
    unmapped = [gid for gid, egi in egi_map.items() if egi is None]

    path.write_text("\n".join(str(e) for e in mapped) + "\n")

    print(f"Written: {path}")
    print(f"  Mapped:   {len(mapped)}")
    if unmapped:
        print(f"  Unmapped: {len(unmapped)}  (no EGI in gene_info_4558)")


def write_symbol(gene_ids: list[str], path: str | Path) -> None:
    """Write gene symbols for the given SORBI gene IDs to a one-per-line text file.

    Skips genes with no symbol mapping and prints a summary.
    """
    path = Path(path)
    path.parent.mkdir(parents=True, exist_ok=True)

    symbol_map = sorbi_to_symbol(gene_ids)
    mapped     = [sym for sym in symbol_map.values() if sym is not None]
    unmapped   = [gid for gid, sym in symbol_map.items() if sym is None]

    path.write_text("\n".join(mapped) + "\n")

    print(f"Written: {path}")
    print(f"  Mapped:   {len(mapped)}")
    if unmapped:
        print(f"  Unmapped: {len(unmapped)}  (no symbol in gene_info_4558)")


# Example: write HIGH-impact gene EGIs and symbols
write_egi(high_genes,    WDIR / "analysis/03_TAA/data/SBC10_HIGH_impact_EGI.txt")
write_symbol(high_genes, WDIR / "analysis/03_TAA/data/SBC10_HIGH_impact_symbol.txt")
```

```python
out = WDIR / "analysis/03_TAA/data/SBC10_HIGH_impact_genes.txt"
out.parent.mkdir(parents=True, exist_ok=True)
out.write_text("\n".join(high_genes) + "\n")
print(f"Written {len(high_genes)} genes → {out}")
```
